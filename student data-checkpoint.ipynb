```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("student.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# drop unnamed column

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

# change weekly study hours

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace(" 5 - 10","6-10")
df.head()
```

# gender distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python

#from the above chart we have analysed that : #the number of females is more than the number of males.
```

#from the above chart we have analysed that :
#the number of females is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Relation between parent education and student score")
plt.show()
```

```python
#from the above chart we have concluded that the education of parent have a good impact on their Scores.
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("Relation between parent MaritalStatus  and student score")
plt.show()
```

```python
#from the above chart we have concluded that the MaritalStatus  of parent have a no  impact on their Scores.
```

# distribution of EthnicGroup

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels= l, autopct="%1.2f%%")
print(mlist)
plt.title("Distribution of Ethnic Groups")
plt.show()
```
